# Generate Paper Assessment Annotations

```python
import pandas as pd
```

```python
df_annotations = pd.read_csv("../data/02_paper_assessment/raw_annotations_paper_assessment.csv")
df_full_text = pd.read_parquet("../data/01_full_text/references_with_full_text.parquet.br")
```

```python
# Add full text by joining on PMID

df_annotations = df_annotations.merge(
    df_full_text[["PMID", "full_text"]],
    on="PMID",
    how="left"
)
```

```python
# Add gt_ prefix to annotated columns

df_annotations = df_annotations.rename(
    columns={col: f"gt_{col}" for col in ["in_scope", "has_repo", "repo_url"]}
)
```

```python
# reorder columns

df_annotations = df_annotations[["PMID", "full_text", "gt_in_scope", "gt_has_repo", "gt_repo_url"]]
```

```python
df_annotations.to_csv("../data/02_paper_assessment/gt_paper_assessment.csv")
```

```python
# Get stats for the RepoAssessment prediction columns
assessment_columns = [
    'is_empty_pred',
    'contains_readme_pred',
    'readme_purpose_and_outputs_pred',
    'contains_requirements_pred',
    'requirements_dependency_versions_pred',
    'contains_license_pred',
    'sufficient_code_documentation_pred',
    'is_modular_and_structured_pred',
    'implements_tests_pred',
    'fixes_seed_if_stochastic_pred',
    'hardware_requirements_pred',
    'contains_link_to_paper_pred',
    'contains_citation_pred',
    'includes_data_or_sample_pred'
]

# Create a summary dataframe with statistics for each column
stats_summary = []

for col in assessment_columns:
    if col in df_pred.columns:
        total_count = len(df_pred)
        non_null_count = df_pred[col].notna().sum()
        null_count = df_pred[col].isna().sum()
        
        # Count True/False values (only for non-null entries)
        true_count = (df_pred[col] == True).sum()
        false_count = (df_pred[col] == False).sum()
        
        # Calculate percentages
        true_pct = (true_count / non_null_count * 100) if non_null_count > 0 else 0
        false_pct = (false_count / non_null_count * 100) if non_null_count > 0 else 0
        
        stats_summary.append({
            'column': col,
            'total_rows': total_count,
            'non_null': non_null_count,
            'null': null_count,
            'true_count': true_count,
            'false_count': false_count,
            'true_pct': f"{true_pct:.2f}%",
            'false_pct': f"{false_pct:.2f}%"
        })

df_stats = pd.DataFrame(stats_summary)
print(df_stats)
```
